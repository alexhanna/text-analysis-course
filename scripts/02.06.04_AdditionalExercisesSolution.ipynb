## Additional Exercises for 02.06: Introduction to Natural Language Processing (NLP) using Python's NLTK

Ex 1:
1. Count the number of stopwords in both novels in the `data` folder.
2. Find the top 20 most frequently used stopwords used in both novels.
3. (Bonus) Make a distribution plot of the top 20 stopword frequencies for both novels.

```python
# Here's the code to load both novels. Make sure you have both text files downloaded in the data folder.
austen_string = open('../data/Austen_PrideAndPrejudice.txt', encoding='utf-8').read()
alcott_string = open('../data/Alcott_GarlandForGirls.txt', encoding='utf-8').read()
```

### 0. Importing required packages and functions to do our analysis

_Note:_ If you're getting an error from the cell below about something like unable to load data, rerun 01.23.02_TechOverview.ipynb

```python
import nltk
from nltk import word_tokenize
import string
from nltk.corpus import stopwords
```

### 1. Tokenizing Text

```python
austen_token = word_tokenize(austen_string)
alcott_token = word_tokenize(alcott_string)
```

### 2. Removing Punctuations

```python
punctuations = list(string.punctuation)
austen_token_clean = [word for word in austen_token if word not in punctuations]
alcott_token_clean = [word for word in alcott_token if word not in punctuations]
```

### 3. Extracting Stopwords

```python
austen_stopwords = [word for word in austen_token_clean if word in stopwords.words('english')]
alcott_stopwords = [word for word in alcott_token_clean if word in stopwords.words('english')]
```

### 4. Lower Casing All Words

```python
austen_stopwords_lc = [word.lower() for word in austen_stopwords]
alcott_stopwords_lc = [word.lower() for word in alcott_stopwords]
```

### 5. Counting the Number of Stopwords

```python
print('Question 1:\n')
print("The number of stopwords in Austen's novel is", str(len(austen_stopwords)))
print("The number of stopwords in Alcott's novel is", str(len(alcott_stopwords)))
```

### 6. Computing the Frequency of All Stopwords

```python
austen_frequency = nltk.FreqDist(austen_stopwords_lc)
alcott_frequency = nltk.FreqDist(alcott_stopwords_lc)

austen_top_20_stopwords = austen_frequency.most_common(20)
alcott_top_20_stopwords = alcott_frequency.most_common(20)
```

```python
print('Question 2:\n')
print("Top 20 most frequently used stopwords in Austen's novel:")
print(austen_top_20_stopwords)

print()
print("Top 20 most frequently used stopwords in Alcott's novel:")
print(alcott_top_20_stopwords)
```

### 7. Plotting the Frequency Distribution of Stopwords

```python
print("Questions 3:\n")
austen_frequency.plot(20, cumulative = False, title = "Stopword Distribution in Austen's Novel")
alcott_frequency.plot(20, cumulative = False, title = "Stopword Distribution in Alcott's Novel")
```
