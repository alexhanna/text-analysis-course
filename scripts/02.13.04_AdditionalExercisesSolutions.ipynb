### Question 1

```python
import pandas
import matplotlib.pyplot as plt

#read in our data
df = pandas.read_csv("../data/childrens_lit.csv.bz2", sep = '\t', encoding = 'utf-8', compression = 'bz2', index_col=0)
```

### Question 2

```python
grouped_year = df.groupby('year')
grouped_year['text'].count().plot(kind = 'bar', title = 'Number of Books per Year')
plt.show()
```

### Question 3

#### 0. Splitting title by white space and count the number of words

*Note:* Here I combined both the function `str.split` and `len` to compute the novel title word length

*Thought Exercise:* Would it still work if I swap `str.split` and `len` around?

```python
df['title length'] = df['title'].apply(str.split).apply(len)
df
```

#### 1. Groupby `year` and compute the average title length

```python
grouped_gender = df.groupby('year').mean()
grouped_gender
```

#### 2. Graph the title length over year

```python
grouped_gender.plot(kind = 'bar', title = 'Average Novel Title Length')
plt.show()
```
